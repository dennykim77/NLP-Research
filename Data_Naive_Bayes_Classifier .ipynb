## Naive Bayes Classifier 

```python
import pandas as pd
```

```python
excelFile = pd.ExcelFile('/Users/taehyung/Desktop/NLP Research/Naive_Bayes_Practice/NLP_data.xlsx')
```

```python
df = pd.read_excel(excelFile, 'clustered sentences (1)')
```

```python
df.head()
```

```python
df.loc[df["clusters"] == 1.0, "Degree of certainty"] = "Uncertain"
df.loc[df["clusters"] == 2.0, "Degree of certainty"] = "NA/Others"
df.loc[df["clusters"] == 3.0, "Degree of certainty"] = "Probable Fact"
df.loc[df["clusters"] == 4.0, "Degree of certainty"] = "Certain Counterfact"
df.loc[df["clusters"] == 5.0, "Degree of certainty"] = "Likely Fact"
df.loc[df["clusters"] == 6.0, "Degree of certainty"] = "Certain Fact"
df.loc[df["clusters"] == 7.0, "Degree of certainty"] = "Possible Fact"
df.loc[df["clusters"] == 8.0, "Degree of certainty"] = "Certain Unkown"
df.loc[df["clusters"] == 9.0, "Degree of certainty"] = "Possible Counterfact"
```

```python
from nltk.tokenize import word_tokenize
import nltk
import numpy as np
```

```python
df1 = pd.DataFrame(df) 
df1 = df1[['sentence', 'Degree of certainty']]
df1
```

```python
train = df1
```

```python
import re
```

```python
train['sentence'] = train['sentence'].astype(str)
train
```

```python
# make all the sentence lowercase for later

train['sentence']= train['sentence'].str.lower()
train.head(15)
```

```python
train['sentence']= train['sentence'].str.replace('` `','')
train.head(15)
```

```python
#######
```

```python
#train1 = re.sub("[^a-zA-Z]", " ", str(train["sentence"]))
#train1
```

```python
# make two columns 'sentence' & 'Degree of certainty' in to list for nltk

train1 = train.reset_index()[['sentence', 'Degree of certainty']].values.astype(str).tolist()
train1
```

### Training Data

```python
all_words = set(word.lower() for sentence in train1
               for word in word_tokenize(sentence[0]))
all_words                                                  # gives you list of all words in the sentences
```

#### all_words is the corpus

```python
#based on created corpus, see if words belong to train1 data

test = [({word: (word in word_tokenize(x[0])) for word in all_words},x[1])
    for x in train1]
test
```

### Naive Bayes Classifier

```python
classifier = nltk.NaiveBayesClassifier.train(test)
classifier.show_most_informative_features()
```

## Testing the Classifier Model

```python
test_sentence = 'Yesterday, I saw a man killing George'
test_sent_features = {word.lower():
                     (word in word_tokenize(test_sentence.lower()))
                     for word in test}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```
